```python
import pymysql
pymysql.install_as_MySQLdb()
import MySQLdb
from sqlalchemy import create_engine
import pandas as pd
```

```python
## Database 접속정보
DB_id = 'root'
DB_pw = '####'
DB_host = 'localhost'
DB_port = '3306'
DB_db = 'data2022'

## DB 연결
engine = create_engine("mysql+mysqldb://"+DB_id+":"+DB_pw+"@"+DB_host+":"+DB_port+"/"+DB_db, encoding='euckr')
con = engine.connect()
```

```python
## Table list
query = 'show tables'
pd.read_sql(query, con)
```

```python
## Read table
data = pd.read_sql_table('tashu_station_info', con)
data
```

```python
data.shape
```

```python
## Copy data frames to database tables
# if_exists = 'replace' : 기존에 있다면 덮어쓰기
data.to_sql(name = 'tashu_station_info_copy', con = con, if_exists = 'replace', index = False)
```
